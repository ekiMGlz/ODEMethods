```python
import odesolver
import numpy as np
import matplotlib.pyplot as plt
from math import *
```

```python
f = lambda t, y : 2*(t+1)*y
soln = lambda t : np.exp(t*(t+2))
```

```python
T,W = odesolver.solve(f, 1, (0,1), 50, method = 'Explicit Euler')

plt.plot(T,W[0],T,soln(T))
plt.title("Explicit Euler")
plt.show()
```

```python
globalError = abs(soln(T) - W[0])
plt.title('Error Global Absoluto - Escala Logaritmica')
plt.scatter(range(len(globalError)), globalError, marker = 'x')
plt.yscale('log')
plt.ylim(1e-4,10)
plt.show()
```
